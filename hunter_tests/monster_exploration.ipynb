```python
# imports
import pandas as pd
import altair as alt

# allow jupytar notebook to create visualizations
alt.renderers.enable("mimetype")

# create a pandas dataframe with the dataframe.csv
df = pd.read_csv('dataframe.csv').drop(columns='Unnamed: 0')
```

```python
# display the head() of df
df.head()
```

```python
# create a altair chart
chart = alt.Chart(df).mark_point().encode(
    x= 'Health',
    y= 'Energy',
    color= 'Rarity'
)

chart.interactive()
```

```python
# create a altair chart
chart = alt.Chart(df).mark_point().encode(
    x= 'Health',
    y= 'Energy',
    color= 'Type'
)

chart.interactive()
```

```python
# create a altair chart
chart = alt.Chart(df).mark_bar().encode(
    x= 'Rarity',
    y= 'Health'
)

chart
```

```python
# create a altair chart
chart = alt.Chart(df).mark_bar().encode(
    x= 'Type',
    y= 'Health'
)

chart
```
